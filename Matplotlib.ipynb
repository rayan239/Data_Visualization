```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x  = np.arange(11)
y = x**2
print(x,y)
```

```python
#scatter plot
plt.scatter(x,y)
```

```python
#plot
plt.plot(x,y)
```

```python
plt.plot(x,y,'r-*')
```

```python
plt.plot(x,y,'r-*')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('plot')
```

```python
plt.plot(x,y,'r-*')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('plot')
plt.xticks(range(0,11))
plt.yticks(range(0,101,10))
plt.show()
```

```python
plt.subplot(1,3,1) # 1 row 3 column
plt.plot(x,y,'r*-')

plt.subplot(1,3,2)
plt.plot(x,y,'g*-')

plt.subplot(1,3,3)
plt.plot(x,y,'b--')
plt.tight_layout()
```

```python
fig1 = plt.figure(figsize=(8,5))
axes = fig1.add_axes([0,0,1,1])
axes.plot(x,y,'r-*')
axes.set_xlabel('x-axis')
axes.set_ylabel('y-axis')
axes.set_title('plot')
axes.set_xticks(range(0,11))
axes.set_yticks(range(0,101,10));

fig1.savefig('fig1.png',dpi=300);
```

```python
fig, axes = plt.subplots(2, 2)   # list1 = ["a", "b"] , figsize=(10,5)
```

```python
# adding nrows and ncols - automaticaly calls the axes based on ncols and nrows
fig, axes = plt.subplots(2, 2)
plt.tight_layout()
```

```python
# We can share x and y axes.
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True, figsize=(8,5))
plt.tight_layout() # for auto adjustment

axes[0][0].plot(x, y, 'r-*')   # abc = [2, 4, 6, 7]
axes[0][0].set_xlabel('X Label') # Notice the use of set_ to begin methods
axes[0][0].set_ylabel('y Label')
axes[0][0].set_title('Figure Title')

axes[1,1].scatter(x, y)
```

```python
# Creating an empty canvas
fig3 = plt.figure(figsize=(10,5))

# # Creating main axes - values b/w 0 & 1
axes_main = fig3.add_axes([0, 0, 1, 1])
# Creating inset axes - values b/w 0 & 1
axes_inset = fig3.add_axes([.3, .5, .3, .4])


axes_main.plot(x, y, 'g-o')
axes_main.set_xlabel('X_label_main')
axes_main.set_ylabel('Y_label_main')
axes_main.set_title('Title_main')

axes_main.set_facecolor("yellow")

axes_inset.plot(y, x, 'r')
axes_inset.set_xlabel('X_label_inset')
axes_inset.set_ylabel('Y_label_inset')
axes_inset.set_title('Title_inset')

#axes_inset.set_facecolor("green")

fig3.savefig("filename.jpg", dpi=200, bbox_inches ='tight')
```

```python
print(x)
print(x**2)
print(x**3)
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x, x**2, label="x**2")
axes.plot(x, x**3, label="x**3")

axes.legend(loc=2)
plt.show()
```

```python
fig, axes = plt.subplots()
axes.plot(x, x+1, 'r.-') # red line with dots
axes.plot(x, x+2, 'g--') # green dashed line
```

```python
fig, axes = plt.subplots()

axes.plot(x, x+1, color="blue", alpha=.5) # half-transparant
axes.plot(x, x+2, color="#8B008B")        # RGB hex code
axes.plot(x, x+3, color="#FF8C00")        # RGB hex code
```

```python
fig, axes = plt.subplots(figsize=(8,6))
axes.plot(x, y, color="purple", lw=3, ls='--', marker='s', markersize=8, markerfacecolor="yellow",
          markeredgewidth=3, markeredgecolor="red");
#Change parameters e.g. markeredgewidth, markerfacecolor, markeredgecolor, marker etc to see the difference
# let's go the the reference jupyter notebook to see some options!
```

```python
fig, axes = plt.subplots(figsize=(10,8))

axes.plot(x, x+1, color="red", linewidth=0.25)
axes.plot(x, x+2, color="red", linewidth=0.50)
axes.plot(x, x+3, color="red", linewidth=1.00)
axes.plot(x, x+4, color="red", linewidth=2.00)

# possible linestype options -, , -., :, steps
axes.plot(x, x+5, color="green", lw=5, linestyle='-')
axes.plot(x, x+6, color="green", lw=5, ls='-.')
axes.plot(x, x+7, color="green", lw=5, ls=':')

# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
axes.plot(x, x+9, color="blue", lw=3, ls='-', marker='o', markersize=8, markerfacecolor="yellow")
axes.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
axes.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
axes.plot(x, x+12, color="blue", lw=3, ls='--', marker='1', markersize=20)

# marker size and color
axes.plot(x, x+13, color="purple", lw=3, ls='-', marker='o', markersize=2)
axes.plot(x, x+14, color="purple", lw=3, ls='-', marker='o', markersize=4)
axes.plot(x, x+15, color="purple", lw=3, ls='-', marker='o', markersize=8, markerfacecolor="red")
axes.plot(x, x+16, color="purple", lw=3, ls='-', marker='s', markersize=8,
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");
```
